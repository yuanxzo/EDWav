```python
"""
This is a sample example of using 'EDWav' method to evaluate seismic waveform segment data
"""
import EDWav
import obspy

# read test wavefrom data by obspy
st=obspy.read('test.sac')

# using obspy for bandpass filtering
st[0].filter('bandpass',freqmin=0.1,freqmax=5, corners=4, zerophase=True)
st[0].plot()

# evaluate the test wavefrom in frequency range of 0.1-5Hz 
obj=EDWav.EDWav.evaluate(st[0].data,FS=st[0].stats.sampling_rate,FB=[0.1,5],FR=98)
# show result
obj.result()


# A simple interpretation of the results:
# 
# This result shows that in the range of 0.1-5Hz, 
# there are two strongly correlated energy clusters, 0.5-2Hz and 2-5Hz, respectively, 
# indicating that the waveform is not diffuse in this frequency band.

```
